```python
import numpy as numpy
import pandas as pd 
```

```python
df = pd.read_csv('customer_shopping_behavior.csv')
df.head()
```

Checking missing values

```python
df.isnull().sum()
```

**So there are 37 review rating null values and we can fill it with mean/ median
but it affect the result because median is a good choice and if we use median it give average according to all categories but we have to fill null acoording to categroy for that use first groupby it**

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
```

```python
df.isnull().sum()
```

here we change the column name in snake case (all letter is in small case)
and instead of space bteween them we use '_' which help us is SQl and python for analysis easily

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
df.columns
```

# Now by taking single ages for analysis is difficult so we and create a new column in which we can group the ages 

```python
# Create age groups
df['age_group'] = pd.cut(
    df['age'],
    bins=[0, 19, 35, 60, 100],
    labels=['Young', 'Adult', 'Middle Age', 'Senior']
)
df[['age','age_group']].head(10)
```

```python
df.columns
```

# SO there are some columns in the the content is in text form and it hard to do analysis from text so we convernt them into int

```python
# create column purchase_frequency_days

frequency_mapping= {
    'Fortnightly': 14,
    'Weekly' : 7,
    'Monthly': 30,
    'Quarterly': 90, 
    'Bi-weekly': 14, 
    'Annually': 365, 
    'Every 3 Month': 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping).fillna(0).astype(int)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```
